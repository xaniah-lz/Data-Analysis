<h1>1. Theft Analysis</h1>

```python
theft1 = theft[theft['Year'] >= 2024]
location = theft1['Location Description'].mode()[0]
theft2 = theft1[theft1['Location Description'] == location]
theft3 = theft2.groupby(['Latitude', 'Longitude']).size().reset_index(name='incident_count')
print("Theft incident count distribution:")
print(theft3['incident_count'].describe())

theft_map = folium.Map(location=[41.7750, -87.7000], zoom_start=12)

heat_data_theft = theft3[['Latitude', 'Longitude', 'incident_count']].values.tolist()
HeatMap(heat_data_theft, min_opacity=0.2, max_val=theft3['incident_count'].max()).add_to(theft_map)
theft_map.save('theft_2024_fixed_scale.html')
print('Data saved to theft_2024_fixed_scale.html')
theft_map
```
