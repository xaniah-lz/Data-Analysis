<h2>IMPORTING LIBRARIES</h2>

```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('datasets\\Global_terrorist.csv', encoding='ISO-8859-1')
```

```python
df
```

<h1>Filtering Data</h1>

```python
filter1 = df[df['region_txt'] == 'Middle East & North Africa']
```

```python
filter2 = filter1[filter1['iyear'] > 2000]
```

```python
filter2['attacktype1_txt'].value_counts()
```

```python
filter3 = filter2[filter2['attacktype1_txt'] == 'Bombing/Explosion']
```

<h1>DATA AGGREGATION</h1>

```python
final_filter = filter3.groupby(['latitude', 'longitude']).size().reset_index(name = 'incident_count')
```

```python
final_filter
```

<h1>Use of FOLIUM FOR GEOSPATIAL</h1>

```python
import folium
from folium.plugins import HeatMap
```

```python
pip install --upgrade pip
```

<h3> Converting Pandas Data Frame into Python List</h3>

```python
final_filter["normalize_data"] = final_filter['incident_count'] - final_filter['incident_count'].min() / final_filter['incident_count'].max() - final_filter['incident_count']

heatmap_data = final_filter[['latitude', 'longitude', 'normalize_data']].values.tolist()
```

<h3> Creating a base map </h3>

```python
my_map = folium.Map(location = [30, 40], zoom_start = 6)
```

<h3> Add Heatmap Layer </h3>

```python
HeatMap(heatmap_data).add_to(my_map)
```

```python
my_map.save('terrorism_bombing.html')
print("data save to terrorism_bombing.html")
```

```python
my_map
```

```python
map1 = folium.Map(location=[20, 0], zoom_start=2)
HeatMap(heatmap_data).add_to(map1)
map1.save('map1_global_overview.html')
print("data saved to map1_global_overview.html")
map1
```

<h1>Insight no.1 </h1>
<h3>This map shows how bombing attacks from terrorist groups have affected many parts of the world. 
It highlights that terrorism is not just a local problem, but a global issue that has touched dozens of countries over the years.</h3>

```python
map2 = folium.Map(location=[20, 0], zoom_start=2)
HeatMap(heatmap_data, radius=15, blur=10).add_to(map2)
map2.save('map2_regional_clusters.html')
print("data saved to map2_regional_clusters.html")
map2
```

<h1>Insight no.2 </h1>
<h3>Bombings tend to happen in groups, not just one place. This tells us that once violence starts in a region,
