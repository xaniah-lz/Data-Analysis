<h1>1. Income Distribution by Region (Bar Plot)

```python
income_by_region = df.groupby('Region')['Total Household Income'].mean().reset_index()

sns.barplot(data=income_by_region, x='Region', y='Total Household Income', palette='coolwarm')

plt.xticks(rotation=45)
plt.title("Average Total Household Income by Region", fontsize=16)
plt.xlabel("Region")
plt.ylabel("Average Total Household Income")
plt.tight_layout()
plt.gcf().set_size_inches(13, 11)
plt.show()
```

<h1>Regional income disparities are clearly evident. Urbanized and economically active regions report significantly higher average household incomes, while more rural or agricultural regions lag behind. This highlights geographic inequality, potentially tied to employment opportunities, infrastructure, and access to services.</h1>

<h1>2. Income vs. Food Expenditure (Scatter Plot)

```python
sns.scatterplot(data=df, x='Total Household Income', y='Total Food Expenditure', alpha=0.6, color='purple')

plt.title("Income vs. Food Expenditure", fontsize=16)
plt.xlabel("Total Household Income")
plt.ylabel("Total Food Expenditure")
plt.grid(True)
plt.tight_layout()
plt.gcf().set_size_inches(13, 11)
plt.show()
```

<h1>There is a positive correlation between income and food expenditure. As household income increases, so does their spending on food, though there are notable outliers where food expenditures are high even with lower income levels.</h1>

<h1>3. Food Expenditure by Region (Bar Plot)

```python
food_by_region = df.groupby('Region')['Total Food Expenditure'].mean().reset_index()

sns.barplot(data=food_by_region, x='Region', y='Total Food Expenditure', palette='Blues')
plt.xticks(rotation=45)
plt.title("Average Total Food Expenditure by Region", fontsize=16)
plt.xlabel("Region")
plt.ylabel("Average Total Food Expenditure")
plt.tight_layout()
plt.gcf().set_size_inches(13, 11)
plt.show()
```

<h1>While income disparities across regions were evident, we also observe that regions with higher incomes like NCR and CALABARZON also report significantly higher food expenditures. This suggests that food expenditure is, to some extent, influenced by income levels.</h1>

<h1>4. Appliance Ownership by Income Quartile (Grouped Bar Chart)

```python
# Create income quartiles
df['Income_Quartile'] = pd.qcut(df['Total Household Income'], 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

appliances = ['Number of Refrigerator/Freezer', 'Number of Washing Machine', 'Number of Airconditioner']
appliance_ownership = df.groupby('Income_Quartile')[appliances].mean()

appliance_ownership.plot(kind='bar', colormap='Set2')
plt.title("Average Appliance Ownership by Income Quartile")
plt.xlabel("Income Quartile")
plt.ylabel("Average Number Owned")
plt.xticks(rotation=0)
plt.legend(title="Appliance")
plt.tight_layout()
plt.grid(True)
plt.gcf().set_size_inches(13, 11)
plt.show()
```

<h1>Appliance ownership strongly correlates with income. Higher-income households (Q3, Q4) own more refrigerators, washing machines, and air conditioners, reflecting better living conditions, while lower-income households own fewer appliances.</h1>

<h1>5. Food Expenditure vs. Income by Region (Scatter Plot with Trendlines)

```python
sns.scatterplot(data=df, x='Total Household Income', y='Total Food Expenditure', hue='Region', alpha=0.6, palette='tab10')
sns.regplot(data=df[df['Region'] == 'NCR'], x='Total Household Income', y='Total Food Expenditure', scatter=False, label='NCR Trend', color='red')
sns.regplot(data=df[df['Region'] == 'CAR'], x='Total Household Income', y='Total Food Expenditure', scatter=False, label='CAR Trend', color='blue')
plt.title("Food Expenditure vs. Income by Region (NCR & CAR Trends Highlighted)")
plt.xlabel("Total Household Income")
plt.ylabel("Total Food Expenditure")
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.gcf().set_size_inches(13, 11)
plt.show()
```

<h1>In high-income regions like NCR, food expenditure grows steadily with income. In contrast, in rural regions like CAR, food expenditure increases more slowly, highlighting economic disparities.</h1>

<h1>6. Ownership of Motor Vehicles by Region (Grouped Bar Chart)

```python
vehicles = ['Number of Car, Jeep, Van', 'Number of Motorcycle/Tricycle', 'Number of Motorized Banca']
vehicles_by_region = df.groupby('Region')[vehicles].mean()

vehicles_by_region.plot(kind='bar', colormap='Set3')
plt.title("Average Ownership of Motor Vehicles by Region")
plt.xlabel("Region")
plt.ylabel("Average Number Owned")
plt.xticks(rotation=45)
plt.legend(title="Vehicle Type")
plt.tight_layout()
plt.grid(True)
plt.gcf().set_size_inches(13, 11)
plt.show()
```

<h1>Motor vehicle ownership patterns vary significantly by region. Urban regions tend to have higher ownership of cars, jeeps, and vans, while motorcycles and tricycles dominate in less urbanized areas. Interestingly, coastal and island regions show relatively higher ownership of motorized bancas, indicating adaptation to local geography and transport needs.

<h1>7. Stove with Oven/Gas Range Ownership vs. Income (Bar Plot)

```python
kitchen_df = df.copy()
kitchen_df['Stove_Owned'] = kitchen_df['Number of Stove with Oven/Gas Range']

sns.barplot(data=kitchen_df, x='Stove_Owned', y='Total Household Income', estimator=np.mean, palette='YlOrBr')
plt.title("Average Income by Number of Gas Ranges/Stoves with Oven")
plt.xlabel("Number of Stove with Oven/Gas Range")
plt.ylabel("Average Household Income")
plt.grid(True)
plt.gcf().set_size_inches(13, 11)
plt.tight_layout()
plt.show()
```

<h1>8. Asset Ownership Heatmap by Income Bracket (heatmap)

```python
# Create income brackets
df['Income Bracket'] = pd.qcut(df['Total Household Income'], q=4, labels=['Low', 'Mid-Low', 'Mid-High', 'High'])

# Select asset columns
asset_cols = [
    'Number of Cellular phone',
    'Number of Personal Computer',
    'Number of Car, Jeep, Van',
    'Number of Motorcycle/Tricycle',
    'Number of Stove with Oven/Gas Range'
]
asset_summary = df.groupby('Income Bracket')[asset_cols].mean()

sns.heatmap(asset_summary, annot=True, cmap='YlGnBu', fmt=".1f")
plt.title("Average Household Assets by Income Bracket")
plt.xlabel("Household Assets")
plt.ylabel("Income Bracket")
plt.gcf().set_size_inches(13, 11)
plt.tight_layout()
plt.show()
```

<h1>Households in higher income brackets tend to own more assets like cars, personal computers, and stoves with ovens. This heatmap clearly shows a positive relationship between income and asset ownership, highlighting economic inequality in household living standards.
